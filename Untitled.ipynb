# Linear regression

```python
from sklearn.linear_model import LinearRegression
lir=LinearRegression()
```

```python
def backtest1(weather, model, predictors, start=3650, step=90): #time series cross validation
    all_predictions = []
    
    for i in range(start, weather.shape[0], step):
        train = weather.iloc[:i,:]#from starting till i'th days(training data) 
        test = weather.iloc[i:(i+step),:]#from i'th till (i+step)days(testing data)
        
        model.fit(train[predictors], train["target"])#(traing of model is performed)
        
    #train[predictors] are the independent variables of traing data
    #train['target'] are the dependent variable training data
    
        preds = model.predict(test[predictors])# testing is performed
        
    #test[predictors] are the independent variable of testing data
    
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()
        
        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions1=backtest1(weather,lir,predictors)
```

```python
predictions1["diff"].plot()
```

# RIDGE REGRESSION

```python
from sklearn.linear_model import Ridge
rr=Ridge(alpha=.1)
```

```python
def backtest2(weather, model_rr, predictors, start=3650, step=90): #time series cross validation
    all_predictions = []
    
    for i in range(start, weather.shape[0], step):
        train = weather.iloc[:i,:]#from starting till i'th days(training data) 
        test = weather.iloc[i:(i+step),:]#from i'th till (i+step)days(testing data)
        
        model_rr.fit(train[predictors], train["target"])#(traing of model is performed)
        
    #train[predictors] are the independent variables of traing data
    #train['target'] are the dependent variable training data
    
        preds = model_rr.predict(test[predictors])# testing is performed
        
    #test[predictors] are the independent variable of testing data
    
        preds = pd.Series(preds, index=test.index)
        combined = pd.concat([test["target"], preds], axis=1)
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()
        
        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions2=backtest2(weather,rr,predictors)
```

# Rolling average of 3 and 14 days

```python
def pct_diff(old, new):
    return (new - old) / old #percentage diff

def compute_rolling(weather, horizon, col):
    label = f"rolling_{horizon}_{col}"
    weather[label] = weather[col].rolling(horizon).mean() #find the mean of the past n days from any date from a perticular coloumn
    weather[f"{label}_pct"] = pct_diff(weather[label], weather[col])
    return weather
    
rolling_horizons = [3, 14]
for horizon in rolling_horizons:
    for col in ["tmax", "tmin", "prcp"]:
        weather = compute_rolling(weather, horizon, col)
```

```python
weather
#pct mean percentage diff
```

```python
weather = weather.iloc[14:,:]#start from 14th number index
weather = weather.fillna(0) #for any null value fill 0
```
